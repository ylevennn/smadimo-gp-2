Ниже выполним подготовку финального, спаршенного датасета. А именно - объединим все полученные части в один большой датасет

Объявим логгер и подключим библиотеки

```python
import pandas as pd
import os
import sys

sys.path.append(os.path.abspath(".."))
from logs.logger import logger

log = logger()
```

Получим имена всех собранных датасетов

```python
# https://sky.pro/wiki/media/rabota-s-fajlami-v-python-kak-poluchit-spisok-vseh-fajlov-v-direktorii/
files_in_data = os.listdir('../../data')
```

```python
games_data_datasets = []
for f in files_in_data:
    if f.startswith('games_data_'):
        games_data_datasets.append(f)

games_data_datasets
```

Откроем каждый файл и объединим в общий датасет

```python
df_parsed = None
for filename in games_data_datasets:
    if df_parsed is None:
        df_parsed = pd.read_excel('../../data/' + filename)
        log.info(f'Добавален первый датасет {filename} | всего строк = {df_parsed.shape[0]}')
        continue

    df_temp = pd.read_excel('../../data/' + filename)
    df_parsed = pd.concat([df_parsed, df_temp], axis=0)
    log.info(f'Добавален датасет {filename} | всего строк в текущем df_parsed = {df_parsed.shape[0]}')

log.info(f'Общий df_parsed собран | всего строк {df_parsed.shape[0]}')
```
